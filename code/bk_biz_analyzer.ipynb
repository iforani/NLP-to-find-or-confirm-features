```python
import pandas as pd
import numpy as np
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
bk_biz = pd.read_csv('../data/brooklyn/reviews/bk_biz_and_rev.csv') #change location since notebook moved
```

```python
def has_dart(review):
    counter = 0
    doc = nlp(review)
    for word in doc:
        if word.lower_ in ["dart"]:
            counter += 1
    return counter
            
```

```python
bk_biz['score'] = bk_biz['reviews'].apply(has_dart)
```

```python
bk_biz
```

```python
bk_biz.loc[bk_biz['score']>0]
```

```python
#same as has_dart(review) with additional POS criteria

def has_dart2(review):
    counter = 0
    doc = nlp(review)
    for word in doc:
        if word.lower_ in ["dart"]:
            if word.pos_ != "VERB":
                counter += 1
    return counter
```

```python
bk_biz['score2'] = bk_biz['reviews'].apply(has_dart2)
```

```python
bk_biz.loc[bk_biz['score2']>0]
```

```python
def matcher_fun(review):
    matcher = Matcher(nlp.vocab)

    #pattern for matcher
    pattern = [{'LOWER': 'dart', 'POS': 'NOUN'}]

    matcher.add('DART', None, pattern)
    
    match = matcher(review)
    
    return match
```
