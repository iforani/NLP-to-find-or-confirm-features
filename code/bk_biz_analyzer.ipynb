```python
import pandas as pd
import numpy as np
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
bk_biz = pd.read_csv('./bk_biz_and_rev.csv') #change location since notebook moved
```

```python
bk_biz.head(3)
```

```python
def has_dart(review):
    counter = 0
    doc = nlp(review)
    for word in doc:
        if word.lower_ in ["dart"]:
            counter += 1
    return counter
            
```

```python
bk_biz['score'] = bk_biz['reviews'].apply(has_dart)
```
